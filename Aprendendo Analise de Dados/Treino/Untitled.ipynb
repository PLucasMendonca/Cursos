Criando variável chamada idn_des para guardar a Series relativa a coluna indice_governanca

Mostrando os valores que estão e não estão em branco

```python
ind_des.isnull()
```

Somando todos os valores que estão em branco

```python
ind_des.isnull().sum()
```

Removendo todos os valores em branco

```python
ind_des.dropna()
```

Verificando se mudamos os valores

```python
ind_des.isnull().sum()
```

Não mudou, por que isso aconteceu? Uma série de funções no pandas não alteram o objeto em questão, porque muitas vezes estamos fazendo transformações temporárias para gerar uma visualização ou análise, para mudar o objeto temos que passar um parâmetro a mais que é o inplace=True.

```python
ind_des.dropna(inplace=True)
```

Verificando novamente

```python
ind_des.isnull().sum()
```
